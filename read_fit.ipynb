```python
import pandas as pd
import fitdecode
from datacyclist.make_dataset import get_dataframes

import os
```

```python
fname = 'data/raw/2023-08-02-16-48-33.fit'
with fitdecode.FitReader(fname) as fit_file:
    for frame in fit_file:
        if isinstance(frame, fitdecode.records.FitDataMessage):
            if frame.name == 'lap':
                res = frame
                break
```

```python
for field in res.fields:
    # field is a FieldData object
    print(field.name)
```

```python
my_fit_files = os.listdir('data/raw/')
my_fit_files.sort()
count = 0

laps_list = []
points_list = []

for file in my_fit_files:
    laps_df, points_df = get_dataframes(f'data/raw/{file}')
    laps_df['activity_no'] = count
    points_df['activity_no'] = count
    laps_list.append(laps_df)
    points_list.append(points_df)
    
    count += 1
```

```python
all_laps = pd.concat(laps_list, ignore_index=True)
all_laps.sample(10)
```

```python
all_points = pd.concat(points_list, ignore_index=True)
all_points.head()
```

```python
all_points.tail()
```

```python
all_laps.to_csv('data/all_laps.csv', index=False)
```

```python
all_points.to_csv('data/all_points.csv', index=False)
```
